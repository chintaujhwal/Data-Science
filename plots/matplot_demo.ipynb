```python
male = tips[tips.sex == 'Male']
male_total = male.groupby('day', as_index = False)['total_bill'].sum()

female = tips[tips.sex == 'Female']
female_total = female.groupby('day', as_index = False)['total_bill'].sum()

plt.plot(male_total['day'], male_total['total_bill'], '--',label = 'male')
plt.plot(female_total['day'], female_total['total_bill'], label = 'female')
plt.legend()
#plt.legend(loc = 'upper right')
plt.xlabel('Day')
plt.ylabel('Total Bill')
plt.title('Sex Vs Total Bill')
#plt.ylim(0,1500)
```

### **Top 10** Countries in **Asia**

#### Method 1

```python
asia_top = countries.loc[countries.region == 'Asia', ['name', 'population', 'area']].sort_values('population', ascending = False).head(10)
asia_top
```

#### Method 2

```python
asia_top = countries.loc[countries.region == 'Asia', ['name', 'population', 'area']].nlargest(10, 'population')
asia_top
```

```python
ax = asia_top['population'].plot.bar()
```

```python
fig,ax = plt.subplots(2,1)
fig.set_size_inches(20,15)

ax[0].bar(asia_top['name'], asia_top['population'])
ax[0].set_xlabel('Country')
ax[0].set_ylabel('Population')
ax[0].set_xticklabels(asia_top.name, rotation = 'vertical')

ax[1].bar(asia_top['name'], asia_top['area'])
ax[1].set_xlabel('Country')
ax[1].set_ylabel('Area')
ax[1].set_xticklabels(asia_top.name, rotation = 90)
```
