```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Tips

```python
tips = sns.load_dataset('tips')
```

### Categorical Plots

#### *boxplot( )*

```python
sns.boxplot(x = 'day', y = 'total_bill', data = tips)
```

```python
sns.boxplot(x = 'day', y = 'total_bill', data = tips, showfliers = False)
```

```python
sns.boxplot(x = 'day', y = 'total_bill', data = tips, hue = 'sex')
```

```python
sns.boxplot(x = 'day', y = 'total_bill', data = tips, hue = 'sex', showfliers = False, showmeans = True)
```

```python
sns.boxplot(x = 'day', y = 'total_bill', data = tips, hue = 'sex', showfliers = False, showmeans = True, meanline = True, width = 0.5)
```

#### *Countplot( )*

```python
sns.countplot(x = 'day', data = tips)
```

```python
sns.countplot(x = 'day', data = tips, hue = 'sex')
```

```python
sns.countplot(x = 'smoker', data = tips, hue = 'day')
```

#### *barplot( )*

#### Method 1

```python
sns.barplot(x = 'day', y = 'total_bill', data = tips, hue = 'size', ci = False, estimator = sum )
```

#### *Catplot( )*

```python
sns.catplot(x = 'day', y = 'total_bill', data = tips)
```

```python
sns.catplot(x = 'day', y = 'total_bill', data = tips, kind = 'bar', ci = False, estimator = sum)
```

```python
sns.catplot(x = 'day', y = 'total_bill', data = tips, col = 'smoker', kind = 'bar', ci = False, estimator = sum)
```

```python
ax = sns.catplot(x = 'size', y = 'total_bill', data = tips, col = 'day', row = 'smoker', kind = 'bar', ci = False)
```

```python
ax = sns.catplot(x = 'day', y = 'total_bill', data = tips, col = 'size', kind = 'box')
```

# Flights

```python
flights = sns.load_dataset('flights')
```

```python
flights.groupby('month')['passengers'].sum()
```

```python
sns.barplot(x = 'month', y = 'passengers', data = flights, ci = False, estimator = sum)
plt.gcf().set_size_inches(15,5)
```

#### *lineplot( )*

```python
sns.lineplot(x = 'month', y = 'passengers', data = flights, ci = False)
plt.gcf().set_size_inches(15,5)
```

```python
sns.lineplot(x = 'year', y = 'passengers', data = flights, ci = False, estimator = sum)
plt.gcf().set_size_inches(15,5)
```

```python
sns.lineplot(x = 'day', y = 'tip', data = tips, hue = 'sex', ci = False)
```

#### *relplot( )*

```python
sns.relplot(x = 'total_bill', y = 'tip', col = 'sex', hue = 'smoker', data = tips)
```

### Multi Plots

#### *pairplot( )*

```python
sns.pairplot(tips, vars = ['total_bill', 'tip'], hue = 'smoker', diag_kind = 'kde')
plt.gcf().set_size_inches(10,5)
```

#### *jointplot( )*

```python
sns.jointplot(x = 'total_bill', y = 'size', data = tips)
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', hue = 'sex', data = tips)
```

### Regression Plots

#### *Regplot( )*

#### *Residplot( )*

#### *FacetGrid( )*

```python
ax = sns.FacetGrid(tips, col = 'day')
ax = ax.map(sns.residplot, 'total_bill', 'tip')
```

```python
ax = sns.FacetGrid(tips, col = 'sex')
ax = ax.map(sns.residplot, 'total_bill', 'tip')
```

### Matrix Plots

```python
sns.heatmap(day_tips)
```

```python
sns.heatmap(day_tips, annot = True)
```

```python
sns.heatmap(day_tips, cmap = 'YlGnBu', annot = True, fmt = '.2f')
```
