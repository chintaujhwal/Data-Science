### Stacking and Splitting

```python
a1 = np.arange(1,5).reshape(2,2)
a1
```

```python
a2 = np.arange(10,16).reshape(2,3)
a2
```

```python
a3 = np.hstack((a1,a2))
a3
```

```python
a4 = np.arange(1,11).reshape(2,5)
a4
```

```python
np.vstack((a3,a4))
```

```python
np.concatenate((a3,a4)) # Vertical Stacking - Axix = 0
```

```python
np.concatenate((a1,a2), axis=1) # Horizontal Stacking - Axix = 1
```

### Image Processing

```python
import matplotlib.pyplot as plt
import matplotlib.image as image
from scipy import misc
```

```python
img = misc.face()
type(img), img.shape, img.size
```

```python
plt.imshow(img)
```

```python
img[0][0] # Colors at index (0,0) --> RGB values
```

```python
img[0,0,1] # Green value at index (0,0)
```

```python
new_image = img.copy()
```

```python
plt.imshow(new_image)
```

```python
new_image[:200, :200, :] = [255, 255, 255]
plt.imshow(new_image)
```

```python
new_image[-200:, -200:, :] = [255,0,0]
plt.imshow(new_image)
```

```python
face = img[200:500, 400:900]
plt.imshow(face)
```

```python
new = img.copy()
n1,n2 = np.vsplit(new, (400,))
```

```python
plt.imshow(n1)
```

```python
plt.imshow(n2)
```

```python
plt.imshow(np.vstack((n1,n2)))
```

```python
(img[:,:,1] == 0).sum() # How many pixels have blue color set to 0
```

```python
i1 = img.copy() # Set red part of the image to 0
i1[:,:,0] = 0 
plt.imshow(i1)
```

```python
i2 = img.copy() # Set red and blue part of the image to 0
i2[:,:,0] = 0 
i2[:,:,2] = 0 
plt.imshow(i2)
```
