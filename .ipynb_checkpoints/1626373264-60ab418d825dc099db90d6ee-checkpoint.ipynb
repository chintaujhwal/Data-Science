```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def linear_regression(x,y,epoch,treshold,lr):       
    n,m = x.shape                            
    betahat = np.zeros((m,1))
    costs = [0]
    for i in range(epoch):
        yhat = x.dot(betahat)
        cost = np.sum(np.square(yhat-y))/2*n
        costs.append(cost)
        differential = x.T.dot((yhat-y))/n
        betahat -= lr*differential
        if abs(cost-costs[-2])<treshold:
            break
    costs.pop(0)
    return costs
```

```python
def random(n,m,noise):
    beta = np.random.uniform(-5,5,(m+1,1))
    x = np.random.uniform(-5,5,(n,m+1))
    x[:,0]=[1]*n
    y = x.dot(beta) + np.random.normal(0,noise,(n,1))
    return x,y,beta
```

```python
x,y,beta=random(20,10,0.2)
cost=linear_regression(x,y,epoch=100,treshold=0.00001,lr=0.05)
print(cost)
```

```python
x,y,beta=random(5,10,0.2)
cost=linear_regression(x,y,epoch=100,treshold=0.00001,lr=0.05)
print(cost)
```
